```python
#cleaning Data
df = pd.read_csv("raw_data.csv")
```

```python
df
```

```python
df.isnull()
df.isna() #same as isnull
```

```python
df.isnull().sum() #used to know which column has how many null value

df.dropna() #delete the raw containing emty entry in it
df.dropna(axis=1) # drop coluumn containing empty value in it

df.fillna(0) # fill all null value with 0 or any value within that parenthesis

age_mean= df["age"].mean()
df["age"].fillna(age_mean) #this won't change the data so we have to use copy of the data and then we have to apply this

cleaned_data=df.copy() # by this we can have original and cleaned data with us
age_mean= df["age"].mean()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)

cleaned_data
```
