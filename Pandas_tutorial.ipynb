```python
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#Access the values in series done by 2 methods (index ans label)
#index
print(s[0])
print(s[3])

#label
```

```python
s=pd.Series([28,23,26,25],index=["Steve","Eleven","Lucus","Dustin"])
print(s)

print(s["Dustin"])
print(s["Steve"])
 #print(s[0])  gives warning to use iloc 
print(s.index)
```

```python
#Properties 
#homogenous
s=pd.Series([28,23,26,25,"Mike"],index=["Steve","Eleven","Lucus","Dustin",25])
print(s)

print(s["Dustin"])
print(s["Steve"])
```

```python
#properties
#vectorization 
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([3,2,3,4,5])
print(s1+s2)

#Mutable values
s1[0]=65
print(s1)

#size of series is immutable
changed_s1=s1.drop(0)

print(s1)
print(changed_s1)

```
